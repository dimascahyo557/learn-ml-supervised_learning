```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

### Shortcut

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1)
plt.plot(X, y2)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r', linewidth=3)
# plt.plot(X, y2, 'b--')
plt.plot(X, y2, 'bo', markersize=2)
# plt.ylim(-2, 2)
plt.xlim(0, 15)

plt.axis('equal')
plt.title("Sin dan Cos")
plt.xlabel('X')
plt.ylabel('y')
```

```python
# Tanpa shortcut
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1, 'r', linewidth=3)
ax.plot(X, y2, 'bo', markersize=2)
ax.set_ylim(-2, 2)
ax.set_xlim(0, 15)
ax.set_title("Sin dan Cos")
ax.set_xlabel("X")
ax.set_ylabel("y")

ax.axis('equal')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1, 'r', linewidth=3, label="Sin")
ax.plot(X, y2, 'bo', markersize=2, label="Cos")
ax.set_ylim(-2, 2)
ax.set_xlim(0, 15)
ax.set_title("Sin dan Cos")
ax.set_xlabel("X")
ax.set_ylabel("y")
ax.axis('equal')
ax.legend()
```

# Scatter Plot

```python
df = pd.read_csv('./data/circle_data.csv', sep=';')
df.head()
```
