### Penggunaan HP Berdasarkan Usia

```python
dailyUsageByAge = df.Daily_Usage_Hours.groupby(df.Age).mean()

X = dailyUsageByAge.keys()
y = dailyUsageByAge.values
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(X, y)
ax.set_title("Penggunaan HP Berdasarkan Usia")
ax.set_xlabel("Usia (tahun)")
ax.set_ylabel("Waktu (jam)")
# ax.set_ylim(np.min([y.min() - 1, 0]), y.max() + 1)
```

### Pengaruh waktu penggunaan terhadap jam tidur

```python
sleepHoursByUsageHours = df.Sleep_Hours.groupby(df.Daily_Usage_Hours).mean()

X = sleepHoursByUsageHours.keys()
y = sleepHoursByUsageHours.values
```

```python
fig = plt.figure()
ax = plt.axes()
ax.plot(X, y)
ax.set_title('Pengaruh waktu bermain HP terhadap jam tidur')
ax.set_xlabel('Lama Penggunaan (jam)')
ax.set_ylabel('Jam Tidur (jam)')
```
